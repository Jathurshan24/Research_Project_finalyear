```python
from PIL import Image
import os
import shutil


resize_method = Image.ANTIALIAS
    #Image.NEAREST)  # use nearest neighbour
    #Image.BILINEAR) # linear interpolation in a 2x2 environment
    #Image.BICUBIC) # cubic spline interpolation in a 4x4 environment
    #Image.ANTIALIAS) # best down-sizing filter


max_height= 40
max_width= 40
extensions= ['JPG']

#method for image resize
def adjusted_size(width,height):
    if width>max_width or height>max_height:
        if width>height:
            return max_width, max_height
        else:
            return max_width, max_height
    else:
        return width,height
    
#image rename and save resized images
for root, dirs, files in os.walk("output1/val"):
#     if not files:
#         continue
#     prefix = os.path.basename(root)
#     for f in files:
#         os.rename(os.path.join(root, f), os.path.join(root, "{}_{}".format(prefix, f)))
        
        if __name__ == "__main__":
            for f in os.listdir(root):
                if os.path.isfile(os.path.join(root,f)):
                    f_text, f_ext= os.path.splitext(f)
                    f_ext= f_ext[1:].upper()
                    if f_ext in extensions:
                        print (f)
                        image = Image.open(os.path.join(root,f))
                        width, height= image.size
                        image = image.resize(adjusted_size(width, height))
                        image.save(os.path.join(root,f))
                        
```
